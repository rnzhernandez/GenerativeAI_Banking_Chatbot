# **Importing Data**

```python
from langchain.document_loaders import TextLoader
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import CharacterTextSplitter
```

```python
loader = PyPDFLoader("/content/Investment_Advising_Guide.pdf")
documents = loader.load()
```

```python
text_splitter = CharacterTextSplitter(chunk_size=250, chunk_overlap=0)
dataset = text_splitter.split_documents(documents)
dataset
```

```python
print(dataset[3].page_content)
```

# **Creation of Vector Database using Embeddings**

```python
from langchain.embeddings import OpenAIEmbeddings
embeddings = OpenAIEmbeddings()
```

```python
text = "Investment advising is a crucial service for clients seeking to grow and protect their wealth."
vector = embeddings.embed_query(text)
vector
```

# **Chroma Vector Database**

```python
from langchain.vectorstores import Chroma
vectorstore = Chroma.from_documents(dataset, embeddings)
```

# **Creation of Vector Store Retriever**

```python
top_k = 5
retriever = vectorstore.as_retriever(search_kwargs={"k": top_k})
```

```python
query_text= "Can you explain the difference between value and growth investing?"
```

```python
relevant_chunks = retriever.get_relevant_documents(query_text)

for idx, chunk in enumerate(relevant_chunks, start=1):
    print(f"Chunk {idx}:\n{chunk.page_content}\n")
```

# **Retrieval Augmented Generation**

```python
query = "Can you explain the difference between value and growth investing?"

vectorstore.similarity_search(query, k=3)
```

```python
def augment_prompt(query: str):
    retriever = vectorstore.similarity_search(query, k=3)

    source_knowledge = "\n".join([x.page_content for x in retriever])
    augmented_prompt = f"""Using the contexts below, answer the query.

    Contexts:
    {source_knowledge}

    Query: {query}"""
    return augmented_prompt
```

```python
print(augment_prompt(query))
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import AIMessage, HumanMessage
import gradio as gr
from fuzzywuzzy import fuzz

def is_greeting(message):
    greetings = ["hi", "hello", "hey", "good morning", "good afternoon", "good evening"]
    return any(fuzz.ratio(message.lower(), greeting) > 80 for greeting in greetings)

def is_investment_related(message):
    investment_keywords = [
        "invest", "portfolio", "stock", "bond", "finance", "market", "return",
        "asset", "risk", "capital", "allocation", "advisor", "investment",
        "tax efficiency", "financial goals", "real estate", "commodities",
        "inflation", "cryptocurrencies", "dollar-cost averaging", "growth investing",
        "value investing", "retirement", "wealth growth"
    ]
    return any(keyword in message.lower() for keyword in investment_keywords)

def predict(message, history):
    if is_greeting(message):
        return "Hello! I'm Finley, your digital assistant for investment advice. How can I assist you today?"

    if not is_investment_related(message):
        return (
            "I'm Finley, and I specialise in investment advising. My responses are focused on investment-related topics "
            "such as portfolio management, investment strategies, risk management, and financial planning. "
            "Please ask me questions within these areas."
        )

    retrieved_docs = vectorstore.similarity_search(message, k=3)
    retrieved_text = "\n\n".join([doc.page_content for doc in retrieved_docs])

    if not retrieved_text.strip():
        return (
            "I'm Finley, an AI specialised in investment advising. My responses are focused on investment-related "
            "topics. How can I assist you within this area?"
        )

    history_langchain_format = []
    for human, ai in history:
        history_langchain_format.append(HumanMessage(content=human))
        history_langchain_format.append(AIMessage(content=ai))

    augmented_message = f"{message}\n\nContext:\n{retrieved_text}"
    history_langchain_format.append(HumanMessage(content=augmented_message))

    gpt_response = chat(history_langchain_format)
    return gpt_response.content

gr.ChatInterface(predict).launch()
```
